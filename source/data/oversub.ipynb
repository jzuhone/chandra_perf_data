```python
from astropy.io import ascii
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
t = ascii.read("Approved_Requested_Proposals.csv")
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(t["Cycle"], t["Requested Observing Time"]*1.0e-3, '-s', ms=10, lw=3, label="Requested")
ax.plot(t["Cycle"], t["Approved Observing Time"]*1.0e-3, '-s', ms=10, lw=3, label="Approved")
ax.set_xticks(t["Cycle"])
ax.set_xticklabels(t["Cycle"], fontdict={"fontsize": 18})
ax.tick_params(axis="y", labelsize=18)
ax.set_xlabel("Cycle", fontdict={"fontsize": 20})
ax.set_ylabel("Time (Ms)", size=18, labelpad=20)
ax.set_title("Requested and Approved Chandra Observing Time", size=24)
ax.legend(fontsize=20, facecolor='white')
#ax.set_xlabel("Year", size=18, labelpad=20)
#ax.arrow(2024, 0, 0, 80, head_width=1.0, head_length=20, fc='C0', ec='C0')
#ax.text(2015, 520, f"{t['papers'].sum()} total papers and counting", bbox=dict(facecolor='white', alpha=1.0), size=22)
fig.savefig("../images/requested_and_approved_time.png", bbox_inches="tight")
```
