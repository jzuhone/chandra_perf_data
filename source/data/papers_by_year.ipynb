```python
from astropy.io import ascii
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
t = ascii.read("Papers_by-year.txt", names=["year", "papers", "papers2"],
               data_end=-1, format="no_header")
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.bar(t["year"], t["papers"])
ax.set_xticks(t["year"])
ax.set_xticklabels(t["year"], rotation=90, fontdict={"fontsize": 18})
ax.tick_params(axis="y", labelsize=18)
ax.set_title("Refereed Chandra Science Papers by Publication Year", size=24)
ax.set_xlabel("Year", size=18, labelpad=20)
ax.arrow(2024, 0, 0, 80, head_width=1.0, head_length=20, fc='C0', ec='C0')
ax.text(2015, 520, f"{t['papers'].sum()} total papers and counting", bbox=dict(facecolor='white', alpha=1.0), size=22)
fig.savefig("../images/referred_papers_mar2024.png", bbox_inches="tight")
```
