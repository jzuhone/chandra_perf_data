```python
from astropy.io import ascii
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
t = ascii.read('uniquePIsCoIs.txt')
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(t["Cycle"], t["DComb"], marker="o", markersize=10, linestyle="-", color="C0")
ax.set_title("Unique Investigators by Cycle", size=24)
ax.set_xticks(t["Cycle"])
ax.set_xticklabels(t["Cycle"],fontdict={"fontsize": 18})
ax.tick_params(axis="y", labelsize=18)
ax.set_ylabel("Cumulative Number", size=18)
ax.set_xlabel("Cycle", size=18)
fig.savefig("../images/investigators.png", bbox_inches="tight")
```
