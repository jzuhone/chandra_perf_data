```python
from astropy.io import ascii
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
import numpy as np
```

```python
t = ascii.read('uniquePIsCoIs.txt')
```

```python
np.diff(t["DComb"]).mean(), np.diff(t["DComb"][-5:]).mean()
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(t["Cycle"], t["DComb"], marker="o", markersize=10, linestyle="-", color="C0")
ax.set_title("Cumulative Number of Unique Investigators over Time", size=24)
ax.set_xticks(t["Cycle"])
ax.set_xticklabels(t["Cycle"],fontdict={"fontsize": 18})
ax.tick_params(axis="y", labelsize=18)
ax.set_ylabel("Cumulative Number", size=18)
ax.set_xlabel("Cycle", size=18)
ax.set_ylim(0, None)
ax.text(1, 5000, "Average added each year after the first = 191", bbox=dict(facecolor='white', alpha=1.0), size=22)
ax.text(1, 4500, "Average added each year in last five years = 186", bbox=dict(facecolor='white', alpha=1.0), size=22)
fig.savefig("../images/investigators.png", bbox_inches="tight")
```
